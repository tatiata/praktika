```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
titanic=pd.read_csv('titanic.csv')
```

# **Box plots**

```python
sns.boxplot(titanic['Age'])
```

```python
sns.boxplot(x='Age', data=titanic)
```

**Kernel density estimate**

```python
sns.kdeplot(titanic['Age'])
```

# **Violin plot**

```python
sns.violinplot(x='Pclass', y='Age', data=titanic)
```

# **Count plots**

```python
sns.countplot(titanic['Survived'])
```

# **Heatmap**

```python
sns.heatmap(titanic.corr(),annot=True,cmap='YlGnBu')
```

# **Facet grid**

```python
g = sns.FacetGrid(titanic, col='Sex', row='Survived')
g.map(plt.hist, 'Age')
```

# **Regplot**

```python
sns.regplot(x='Age', y='Fare', data=titanic)
```

# **lmplot**

```python
sns.lmplot(x='Age', y='Fare', data=titanic, col='Sex')
```

# **Strip plot**

```python
sns.stripplot(x='Survived', y='Age', data=titanic)
```

# **Swarm plot**

```python
sns.swarmplot(x='Survived', y='Age', data=titanic)
```

ти типы графиков можно комбинировать с другими. Лучше всего это делать с violinplot.

```python
sns.violinplot(x='Survived', y='Age', data=titanic, palette='rainbow')
sns.swarmplot(x='Survived', y='Age', data=titanic, color='black', size=3)
```

# **Pair plot**

```python
sns.pairplot(data=titanic)
```

**в цветовой палитре**

```python
sns.pairplot(titanic, hue='Sex', palette='Set1')
```

# **Joint plot**

```python
sns.jointplot(x='Fare', y='Age', data=titanic)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
```

```python
data=load_iris()
data
```

```python
iris=pd.DataFrame(data=data.data, columns=data.feature_names)
iris
iris['species']=data.target
iris['species']=iris['species'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
iris['species'].value_counts().plot(kind='bar')
fig=plt.figure(figsize=(10,5))
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig,a = plt.subplots(1,2)

ax1=fig.add_subplot(111)
ax2=fig.add_subplot(121)

iris.plot(kind='scatter', x='petal length (cm)', y='petal width (cm)', ax=ax1)

ax1.scatter(x='petal length (cm)', y='petal width (cm)', data=iris)

ax2.scatter(x='sepal length (cm)', y='sepal width (cm)', data=iris)

plt.show()
```
