# Matplotlib обьектно - ориентированный интерфейс 

```
import matplotlib.pyplot as plt
import numpy as np 
import math
x = linspace(-3, 3, 30)
y = x**2
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("Тригонометрическая фукция")
ax.set_xlabel('угол')
ax.set_xlabel('sin()')
plt.show
```

```
import matplotlib.pyplot as plt
y = [1, 4, 9, 16, 25, 36, 49, 64]
x1 = [1, 16, 30, 30, 42, 55, 68, 77, 88]
x2 = [1, 6, 12, 18, 28, 40, 52, 65]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1,y,'ys-')
l2 = ax.plot(x2, y, 'go--')
ax.legend(labels = ('tv', 'Smartphone'), loc = 'lower right')
ax.set_title("Тригонометрическая фукция")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np 
import math
x = np.arange(0, np.pi*2, 0.05)
fig=plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3])
y = np.sin(x)
axes1.plot(x, y, 'b')
axes2.plot(x, np.cos(x), 'r')
axes1.set_title('sin')
axes2.set_title('cosin')
plt.show()
```

# Функция Matplotlib - subplots

```
import matplotlib.pyplot as plt
fig,a = plt.subplots(2,2)
import numpy as np 
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('sguare')
a[0][1].plot(x, np.sqrt(x))
a[0][1].set_title('sguare root')
a[1][0].plot(x, np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x, np.log10(x))
a[1][1].set_title('log')
plt.show
```

# Функция Matplotlib - subplot2grid

```
import matplotlib.pyplot as plt
a1 = plt.subplot2grid((3,3),(0,0), colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0), colspan = 2, rowspan = 2)
import numpy as np 
x = np.arange(1,10)
a2.plot(x, x*x, 'r.')
a2.set_title('square')
a1.plot(x, np.exp(x), 'g')
a1.set_title('exp')
a3.plot(x, np.log(x), 'y.')
a2.set_title('log')
plt.tight_layout()
plt.show()
```

# Matplotlib - Bar

```
import matplotlib.pyplot as plt
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
ax.bar(langs,students)
plt.show()
```

```
import numpy as np 
import matplotlib.pyplot as plt
data = [[30,25,50,20],[40,23,51,17], [35,22,45,19]]
X = np.arange(4)
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)
```

# Matplotlib  - круговая диаграма

```
import matplotlib.pyplot as plt
import numpy as np 
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
ax.pie(students, labels = langs, autopct='%1.2f%%')
plt.show()
```
